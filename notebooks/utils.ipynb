```python
import zipfile
from pathlib import Path

zip_path = Path("../data/raw/ESCO dataset - v1.2.1 - classification - en - csv.zip")
out_dir  = Path("../data/raw/esco_v1.2.1_en_csv")
out_dir.mkdir(parents=True, exist_ok=True)

with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(out_dir)

print("Extracted to:", out_dir)
print("Some files:", sorted([p.name for p in out_dir.glob("*.csv")])[:20])
```
